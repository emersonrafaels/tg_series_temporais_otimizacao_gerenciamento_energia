# IMPORTANDO AS BIBLIOTECAS

```python
import re

import pandas as pd
import numpy as np
import bq_helper
from bq_helper import BigQueryHelper
```

```python
def convert_fahrenheit_to_celsius(value_fahrenheit):
    
    """
    
        REALIZA A CONVERSÃO DE UM VALOR
        DA ESCALA FAHRENHEIT PARA CELSIUS.
        
        # Arguments
            value_fahrenheit   - Required: Valor na escala fahrenheit (Integer | Float)
        
        # Returns
            value_celsius      - Required: Valor na escala celsius (Integer | Float)
    
    """
    
    value_celsius = (value_fahrenheit - 32) * 5/9
    return value_celsius
```

```python
def convert_milibar_to_pascal(value_milibar):
    
    """
    
        REALIZA A CONVERSÃO DE UM VALOR
        DA ESCALA FAHRENHEIT PARA CELSIUS.
        
        # Arguments
            value_milibar      - Required: Valor na escala milibar (Integer | Float)
        
        # Returns
            value_pascal       - Required: Valor na escala Pascal (Integer | Float)
    
    """
    
    value_pascal = value_milibar*100
    return value_pascal
```

```python
def convert_knots_to_ms(value_knots):
    
    """
    
        REALIZA A CONVERSÃO DE UM VALOR
        DA ESCALA FAHRENHEIT PARA CELSIUS.
        
        # Arguments
            value_knots        - Required: Valor na escala knots (Integer | Float)
        
        # Returns
            value_ms           - Required: Valor na escala metros por segundo (m/s) (Integer | Float)
    
    """
    
    value_ms = value_knots*0.514444
    return value_ms
```

```python
def convert_inch_to_meters(value_inch):
    
    """
    
        REALIZA A CONVERSÃO DE UM VALOR
        DA ESCALA FAHRENHEIT PARA CELSIUS.
        
        # Arguments
            value_inch         - Required: Valor na escala inch (Integer | Float)
        
        # Returns
            value_meters       - Required: Valor na escala metros (m) (Integer | Float)
    
    """
    
    value_meters = value_inch*0.0254
    return value_meters
```

# INSTANCIANDO O DATASET DESEJADO

## O dataset está no bigquery do GCP

```python
ghcn_daily = bq_helper.BigQueryHelper(active_project="bigquery-public-data", 
                                      dataset_name="noaa_gsod")
```

## OBTENDO A LISTA DE ANOS DISPONÍVEIS

```python
bq_assistant = BigQueryHelper("bigquery-public-data", 
                              "noaa_gsod")
list_tables = bq_assistant.list_tables()
```

## FILTRANDO TABELAS DE ANOS

```python
list_tables_year = [table for table in list_tables if re.search(r'\d{4}', table)]
```

```python
list_tables_year
```

# OBTENDO OS DADOS

```python
# INICIANDO O DATAFRAME
df = pd.DataFrame()

for year in list_tables_year:
    
    try:
        print("ANO ATUAL: {}".format(year))

        QUERY = """
        SELECT
          ghcn.stn,
          ghcn.year,
          ghcn.mo AS month,
          ghcn.da AS day,
          stations.name,
          stations.country,
          stations.state,
          stations.lat,
          stations.lon,
          stations.begin,
          stations.end,
          ghcn.temp,
          ghcn.prcp,
          CAST(ghcn.wdsp AS float64) as wdsp,
          ghcn.sndp
        FROM
          `bigquery-public-data.noaa_gsod.{year}` ghcn
        INNER JOIN
          `bigquery-public-data.noaa_gsod.stations` stations
        ON
          ghcn.stn = stations.usaf
        WHERE
          stations.country LIKE 'BR'
        """.format(year=year)

        df_weather = bq_assistant.query_to_pandas(QUERY)

        df = pd.concat([df, df_weather])
    except Exception as ex:
        print(ex)
```

# PÓS-PROCESSAMENTO

## CONVERSÃO DA TEMPERATURA DE 'FAHRENHEIT (F)' PARA 'CELSIUS (C)'

```python
df["temp_celsius"] = df["temp"].apply(lambda x: convert_fahrenheit_to_celsius(x))
```

## CONVERSÃO DA PRECIPITAÇÃO MÉDIA DE 'MILIBAR (mb)' PARA 'PASCAL (Pa)'

```python
df["prcp"] = df["prcp"].apply(lambda x: convert_milibar_to_pascal(x))
```

## CONVERSÃO DA VELOCIDADE MÉDIA DO VENTO DE 'KNOTS (kn)' PARA 'METROS POR SEGUNDO (m/s)'

```python
df["wdsp"] = df["wdsp"].apply(lambda x: convert_knots_to_ms(x))
```

## CONVERSÃO DA PROFUNDIDADE DA NEVE ACUMULADA DE 'INCH (in)' PARA 'METROS (m)'

```python
df["sndp"] = df["sndp"].apply(lambda x: convert_inch_to_meters(x))
```

### CONVERTENDO VALORES ATÍPICOS PARA NAN

```python
df["prcp"] = df["prcp"].replace(9999.00, np.nan)
```

# EXPORTANDO O RESULTADO

```python
df.to_csv('data_brazil_daily.csv', index=False)
```
